```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../cpp/prueba")
times = np.arange(0,100,0.1)

print(data.shape)

sum_pops = np.mean(data[:,::2], axis=1)
resources = np.mean(data[1:,::2], axis=1)

plt.figure()
plt.plot(times, data[:,0])
plt.xlim(0,10)
plt.show()

plt.figure()
plt.plot(times, sum_pops)
plt.xlim(0,10)
plt.show()
```

```python
plt.figure()

for k in [0.001, 1,2]:
    data = np.loadtxt(f"../cpp/k{k}")
    times = np.arange(0,100,0.1)

    sum_pops = np.mean(data[:,::2], axis=1)

    plt.plot(times, sum_pops, label=k)
plt.legend(loc="best")
#plt.xlim(0,10)
plt.show()



plt.figure()

for k in [1,2]:
    data = np.loadtxt(f"../cpp/k{k}")
    times = np.arange(0,100,0.1)

    resources = np.mean(data[:,1::2], axis=1)

    plt.plot(times, resources, label=k)
plt.legend(loc="best")
#plt.xlim(0,10)
plt.show()
```
